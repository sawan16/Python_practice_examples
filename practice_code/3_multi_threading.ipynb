```
from threading import *

class ThreadDemo(Thread):
  def run(self):
    i = 0
    while i< 10:
      print(current_thread().getName(), 'performing task.....\n')
      i += 1


t1 = ThreadDemo()
t1.start()

for j in range(10):
  print(current_thread().getName(), 'performing task.....\n')
  
```
