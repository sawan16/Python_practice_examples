# Static Variables

```
class A:
  var1 = 'val1'

  def __init__(self):
    A.var2 = 'val2'

  @classmethod
  def m1(cls):
    cls.var3 = 'val3'
    A.var4 = 'val4'
  
  @staticmethod
  def m2():
    A.var5 = 'val5'
  
  def m3(self):
    A.var6 = 'val6'

print(A.__dict__)
A().m3()
print(A.__dict__)
A.m1()
A.m2()
print(A.__dict__)
```

# Methods

```
class Product:

  def __init__(self, name, weight, state):
    self.name = name
    self.weight = weight
    self.state = state

  def product_details(self):
    print('product name:%s weight:%s state:%s'%(self.name, self.weight, self.state))

p1 = Product('Bat', '2.1 Kg', 'solid')
p2 = Product('ABC Juice', '4 L', 'liquid')

p1.product_details()
p2.product_details()
```

```
class Product:

  def __init__(self, name, weight, state):
    self.name = name
    self.weight = weight
    self.state = state

  def set_name(self, name):
    self.name = name

  def product_details(self):
    print('product name:%s weight:%s state:%s'%(self.name, self.weight, self.state))

p1 = Product('Bat', '2.1 Kg', 'solid')
p2 = Product('ABC Juice', '4 L', 'liquid')

p1.set_name('Bat_1')

p1.product_details()
p2.product_details()
```

# Polymorphism

```
class Truck:
  def m(self):
    print('I am truck..')

class Car:

  def m(self):
    print('I am Car..')

class Bike:

  def m(self):
    print('I am Bike..')

list1 = [Truck(), Bike(), Bike(), Car()]

for w in list1:
  print(w.m())
```

# Operator Overloading

```
class RAM:

  def __init__(self, size):
    self.size = size

  def __add__(self, r):
    return self.size+r.size

r1 = RAM(4)
r2 = RAM(8)

r1+r2
```

```
+ ---> object.__add__(self,other)
- ---> object.__sub__(self,other)
* ---> object.__mul__(self,other)
/ ---> object.__div__(self,other)
// ---> object.__floordiv__(self,other)
% ---> object.__mod__(self,other)
** ---> object.__pow__(self,other)
+= ---> object.__iadd__(self,other)
-= ---> object.__isub__(self,other)
*= ---> object.__imul__(self,other)
/= ---> object.__idiv__(self,other)
//= ---> object.__ifloordiv__(self,other)
%= ---> object.__imod__(self,other)
**= ---> object.__ipow__(self,other)
< ---> object.__lt__(self,other)
<= ---> object.__le__(self,other)
> ---> object.__gt__(self,other)
>= ---> object.__ge__(self,other)
```
