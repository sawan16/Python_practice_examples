```
import re

pattern = re.compile('ee')

result = pattern.finditer('wwweetttettteeeteeee')

for r in result:
  print(r)
```

```
import re

result = re.finditer('[a-z]','RnYtt67gG#%uU')  # [^ab] [a-zA-Z] [0-9] [^a-zA-Z0-9]

for r in result:
  print(r)
```

```
import re

result = re.finditer('R{2,4}','RRRtttyyy43RRuuRRRRRaaRiiiiRkkkk')  # R*, R+, R?, R{2}

for r in result:
  print(r)
```

```
import re
s='kgb'
m=re.match(s,"kgbjjuiiid") #iikgbjjuiiid
print(m)
if m!= None:
  print("Start Index:",m.start(), "and End Index:",m.end()) 
```
