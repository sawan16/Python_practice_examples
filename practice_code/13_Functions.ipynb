```
def decor(func):
 def inner(name):
   print("First Decor(decor) Function Execution")
   func(name)
 return inner

def decor1(func):
 def inner(name):
   print("Second Decor(decor1) Execution", len(name))
   func(name)
 return inner

@decor1
@decor
def welcome(name):
 print("Hello",name)

welcome("sawan")
```
