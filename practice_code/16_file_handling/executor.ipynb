```
f=open("file1.txt",'r')
print("File Name: ",f.name)
print("File Mode: ",f.mode)
print("Is File Readable: ",f.readable())
print("Is File Writable: ",f.writable())
print("Is File Closed : ",f.closed)
f.close()
print("Is File Closed : ",f.closed)
```

```
f1 = open('file1.txt')
print(f1.tell())
print(f1.read(5))
print(f1.tell())
```

```
f1 = open('file1.txt')
print(f1.tell())
print(f1.seek(6))
print(f1.tell())
print(f1.read(2))
print(f1.seek(2))
print(f1.tell())
print(f1.seek(2,1))
print(f1.tell())
```

```
import csv

f1 = open('csv_file.csv', 'w')

wtr = csv.writer(f1)
wtr.writerow(['number', 'square'])

for i in range(28):
  wtr.writerow([i,i**2])

f1.close()
```

```
f1 = open('csv_file.csv')

rdr = csv.reader(f1)

data = list(rdr)

for row in data:
  print(row)

```
